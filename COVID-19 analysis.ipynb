# COVID-19 Visualization

Data provided by Johns Hopkins University

### Import libraries

```python
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns
import datetime
```

### Load data

```python
conf_csv = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
deaths_csv = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
recovered_csv = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
```

The data gets updated every day.

## Confirmed COVID - 19 Cases

```python
conf_df = pd.read_csv(conf_csv)
conf_df.head()
```

We have data for the total confirmed cases reported till yesterday across the globe.

```python
conf_df.shape
```

```python
global_cases=conf_df.iloc[:,4:].apply(sum,axis=0)
global_cases.index=pd.to_datetime(global_cases.index)
print("Former confirmed cases across globe : \n", global_cases.head(10))
print("\n Recent confirmed cases across globe :\n", global_cases.tail(10))
```

### Total confirmed cases in China, India and US

```python
china_cases = conf_df[conf_df['Country/Region'] == 'China'].iloc[:,4:].apply(sum,axis=0)
china_cases.index = pd.to_datetime(china_cases.index)
print("Recent Confirmed Cases in China :\n",china_cases.tail())

india_cases = conf_df[conf_df['Country/Region'] == 'India'].iloc[:,4:].apply(sum,axis=0)
india_cases.index = pd.to_datetime(india_cases.index)
print("\n Recent Confirmed Cases in India :\n",india_cases.tail())

us_cases = conf_df[conf_df['Country/Region'] == 'US'].iloc[:,4:].apply(sum,axis=0)
us_cases.index = pd.to_datetime(us_cases.index)
print("\n Recent Confirmed Cases in US :\n",us_cases.tail())
```

The output shows the <b>Recent Confirmed Cases</b> in <b>China, India and US.</b> It is seen that <b>China has less confirmed cases</b> compared to <b>India</b> and <b>US</b>.

####  Confirmed COVID-19 Cases Across Globe

```python
plt.style.use('dark_background')
plt.figure(figsize=(20,6))
plt.title('Total Confirmed COVID-19 Cases Across Globe')
plt.plot(global_cases.index, global_cases, c='orange', linewidth=2, marker='o', markersize=4)
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.show()
```

The above graph shows that there was a drastic increase of Confirmed Cases of COVID-19 in mid-March 2020 across Globe.

#### Confirmed Cases in China

```python
plt.figure(figsize=(20,6))
plt.title('Total Confirmed COVID-19 Cases in China')
plt.plot(china_cases.index, china_cases, c='green', linewidth=2, marker='o', markersize=4)
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.show()
```

The curve flattens after 23 February 2020 as Wuhan (the capital of Hubei in China) was put under lockdown on 23 January 2020. As a result, the spread of the virus slowed down.

#### Confirmed COVID-19 Cases in India

```python
plt.figure(figsize=(20,6))
plt.title('Total Confirmed COVID-19 Cases in India')
plt.plot(india_cases.index, india_cases, c='magenta', linewidth=2, marker='o', markersize=4)
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.show()
```

India started testing in bulk in the last week of March 2020 as the outbreak started increasing in last week of March 2020

#### Confirmed COVID-19 Cases in US

```python
plt.figure(figsize=(20,6))
plt.title('Total Confirmed COVID-19 Cases in India')
plt.plot(us_cases.index, us_cases, c='red', linewidth=2, marker='o', markersize=4)
#plt.xticks(rotation=45)
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.show()
```

US started testing in bulk in mid-March 2020 as the outbreak started increasing around mid-March 2020

#### Confirmed COVID-19 Cases in China and India

```python
plt.figure(figsize=(20,6))
plt.title('Total Confirmed COVID-19 Cases Reported in China and India')
plt.plot(china_cases.index, china_cases, c='orange', linewidth=2, marker='*', markersize=4,label='China')
plt.plot(india_cases.index, india_cases, c='magenta', linewidth=2, marker='o', markersize=4,label='India')
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.legend()
plt.show()
```

The above graph shows comparison between China and India. Total Confirmed COVID-19 Cases decreased in China because of the lockdown imposed at the early stage of the outbreak.

#### Confirmed COVID-19 Cases in China, India and US

```python
plt.figure(figsize=(20,6))
plt.title('Total COVID-19 Cases Reported in India, China and US')
plt.plot(china_cases.index, china_cases, c='orange', linewidth=2, marker='*', markersize=4,label='China')
plt.plot(india_cases.index, india_cases, c='magenta', linewidth=2, marker='o', markersize=4,label='India')
plt.plot(us_cases.index, us_cases, c='red', linewidth=2, marker='^', markersize=4,label='US')
plt.xticks(rotation=45)
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.legend()
plt.show()
```

At the beginning of the last week of March, the total confirmed cases of coronavirus in the USA surpassed the same in China.

<i>We'll visualize the trends starting from March 15, 2020 because that's when the USA started testing in bulk. They have recorded the most number of COVID-19 confirmed cases till date</i>

### Top 5 Countries having most no. of COVID-19 confirmed cases

```python
grouped_conf_df = conf_df.groupby(by='Country/Region',as_index=False).sum()
desc_grp_conf_df = grouped_conf_df.sort_values(by=conf_df.columns[-1],ascending=False)
desc_grp_conf_df.head()
```

```python
plt.figure(figsize=(20,8))
plt.title('Top 5 Countries with Highest COVID-19 Confirmed Cases')
for region in desc_grp_conf_df.head()['Country/Region']:
    plt.plot(global_cases.index[53:],desc_grp_conf_df[desc_grp_conf_df['Country/Region'] == region].iloc[0,56:],'-o',markersize=4,label=region)
plt.legend()
plt.grid(True,linestyle='--',c='grey')
plt.show()
```

### Top 20 Countries having most no. of COVID-19 confirmed cases

```python
last_col = conf_df.columns[-1]
latest_date = datetime.datetime.strptime(last_col, '%m/%d/%y').strftime('%B %d, %Y')
plt.figure(figsize=(15,10))
plt.title('Top 20 Countries with Highest COVID-19 Confirmed Cases')
sns.barplot(desc_grp_conf_df[last_col].head(20),desc_grp_conf_df['Country/Region'].head(20),orient='h')
plt.xlabel(f'Total Confirmed Cases (in millions) as of {latest_date}')
plt.show()
```

The above analysis shows the countries with highest number of confirmed cases across Globe. It is seen that US has the highest number of Confirmed COVID-19 Cases across Globe followed by Brazil and India.

## Recovered COVID - 19 Cases

```python
recovered_df = pd.read_csv(recovered_csv)
grouped_rec_df = recovered_df.groupby(by='Country/Region',as_index=False).sum()
desc_grp_rec_df = grouped_rec_df.sort_values(by=recovered_df.columns[-1],ascending=False)
desc_grp_rec_df.head()
```

```python
china_rec_cases = recovered_df[recovered_df['Country/Region'] == 'China'].iloc[:,4:].apply(sum,axis=0)
china_rec_cases.index = pd.to_datetime(china_rec_cases.index)
print("Recent Recovered Cases in China :\n",china_rec_cases.tail())

india_rec_cases = recovered_df[recovered_df['Country/Region'] == 'India'].iloc[:,4:].apply(sum,axis=0)
india_rec_cases.index = pd.to_datetime(india_rec_cases.index)
print("\n Recent Recovered Cases in India :\n",india_rec_cases.tail())

us_rec_cases = recovered_df[recovered_df['Country/Region'] == 'US'].iloc[:,4:].apply(sum,axis=0)
us_rec_cases.index = pd.to_datetime(us_rec_cases.index)
print("\n Recent Recovered Cases in US :\n",us_rec_cases.tail())
```

### Recovered COVID-19 Cases in China and India

```python
plt.figure(figsize=(20,6))
plt.title('Total Recovered COVID-19 Cases Reported in China and India')
plt.plot(china_rec_cases.index, china_cases, c='orange', linewidth=2, marker='*', markersize=4,label='China')
plt.plot(india_rec_cases.index, india_cases, c='magenta', linewidth=2, marker='o', markersize=4,label='India')
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.legend()
plt.show()
```

### Recovered COVID-19 Cases in China, India and US

```python
plt.figure(figsize=(20,6))
plt.title('Total Recovered COVID-19 Cases Reported in India, China and US')
plt.plot(china_rec_cases.index, china_cases, c='orange', linewidth=2, marker='*', markersize=4,label='China')
plt.plot(india_rec_cases.index, india_cases, c='magenta', linewidth=2, marker='o', markersize=4,label='India')
plt.plot(us_rec_cases.index, us_cases, c='red', linewidth=2, marker='^', markersize=4,label='US')
plt.ylabel('in millions')
plt.grid(True,'major',linestyle='--',c='grey')
plt.legend()
plt.show()
```

### Top 5 Countries having most no. of COVID-19 recovered cases

```python
plt.figure(figsize=(20,8))
plt.title('Top 5 Countries with Highest COVID-19 Recovered Cases')
for region in desc_grp_rec_df.head()['Country/Region']:
    plt.plot(global_cases.index[53:],desc_grp_rec_df[desc_grp_rec_df['Country/Region'] == region].iloc[0,56:],'-o',markersize=4,label=region)
plt.legend()
plt.grid(True,linestyle='--',c='grey')
plt.show()
```

### Top 20 Countries having most no. of COVID-19 recovered cases

```python
plt.figure(figsize=(15,10))
plt.title('Top 20 Countries with Highest COVID-19 Recovered Cases')
sns.barplot(desc_grp_rec_df[last_col].head(20),desc_grp_rec_df['Country/Region'].head(20),orient='h')
plt.xlabel(f'Total Recovered Cases (in millions) as of {latest_date}')
plt.show()
```

The above analysis shows the countries with highest number of recovered cases across Globe. It is seen that Brazil has the highest number of Recovered COVID-19 Cases across Globe followed by US and India.
